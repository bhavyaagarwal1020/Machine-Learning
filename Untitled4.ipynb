```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
from matplotlib.figure import Figure
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
import sklearn.datasets
```

```python
hp_data = sklearn.datasets.fetch_california_housing()
```

```python
df = pd.DataFrame(hp_data.data, columns = hp_data.feature_names)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['price'] = hp_data.target
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
corelation = df.corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(corelation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size':5}, cmap='Blues')
```

```python
X = df.drop(['price'], axis = 1)
```

```python
Y = df['price']
```

```python
print(X)
print(Y)
```

```python
#splitting data into training data and testing data
X_train, X_test, Y_test, Y_train = train_test_split(X, Y, test_size = 0.5, random_state=100)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Model Training
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#accuracy for prediction on training data
train_data_prediction = model.predict(X_train)
```

```python
print(train_data_prediction)
```

```python
#R sqaured error
score1 = metrics.r2_score(Y_train, train_data_prediction)

#mean absolute error
score2 = metrics.mean_absolute_error(Y_train, train_data_prediction)

print("R Sqaured error = ", score1)
print("Mean Absolute error = ", score2)

```

```python
#accuracy for prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
#R sqaured error
score1 = metrics.r2_score(Y_test, test_data_prediction)

#mean absolute error
score2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R Sqaured error = ", score1)
print("Mean Absolute error = ", score2)
```

Visualizing actual price and predicting price

```python
plt.scatter(Y_train, train_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("predicted prices")
plt.title("ACTUAL VS PREDICTED")
plt.show()
```
